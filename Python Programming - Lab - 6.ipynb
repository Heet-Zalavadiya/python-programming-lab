### 17) WAP to create a list of common elements from given two lists.

```python
l1 = list(map(int, input("Enter numbers separated by space: ").split()))
l2 = list(map(int, input("Enter numbers separated by space: ").split()))
l3 = []
for i in l1:
    for j in l2:
        if i == j:
            l3.append(i)
print(list(set(l3)))
```
