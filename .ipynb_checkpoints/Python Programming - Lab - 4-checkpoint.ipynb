### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter a number: "))
flag = 0
for i in range(2,n):
    if n % i == 0:
        flag = 1
if flag == 1:
    print("not prime number")
else:
    print("prime number")
```

### 13) WAP to print sum of digits of given number.

```python
import math
n = int(input("Enter a number: "))
sum = 0
while n!=0:
    sum += n % 10
    n = math.floor(n / 10)
print(sum)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
import math
n = int(input("Enter a number: "))
temp = n
rev = 0
while temp!=0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = math.floor(temp / 10)
if n == rev:
    print("number is palindrom")
else:
    print("number is not palindrom")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
import math
n = int(input("Enter a number: "))
temp = n
count  = 0
sum = 0
while temp!=0:
    digit = temp % 10
    count += 1
    temp = math.floor(temp / 10)
temp = n
while temp!=0:
    digit = temp % 10
    sum += digit ** count
    temp = math.floor(temp / 10)
if sum == n:
    print("number is armstrong")
else:
    print("number is not armstrong")
```

### 16) WAP to print all the perfect numbers between 1 to n.
